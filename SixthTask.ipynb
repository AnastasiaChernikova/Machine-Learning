# Learning Rate Decay
Коефіцієнт навчання є одним з найважливіших гіперпараметрів для налаштування нейронної мережі для досягнення кращого перформансу. Коефіцієнт навчання визначає розмір кроку при кожній ітерації тренувань, рухаючись до оптимальної функції втрат. Як правило, рівень навчання позначається символом α. Значення α визначається в діапазоні між 0 і 1.

Напрямок до оптимальної функції втрат можна знайти, обчисливши градієнт функції втрат. Параметр швидкості навчання визначає, наскільки великим вважається розмір кроку в цьому напрямку.
Нейронна мережа складається з двох процедур, таких як Forward propagation та Back-propagation.

    Forward propagation є також відомий як feedforward, який використовувався для передбачення вихідної змінної.
    Back-propagation method використовується для оновлення ваги та зміщення кожного шару, щоб мінімізувати функцію втрат.

 

В Back-propagation method, параметри ваги та зміщення оновлюються за допомогою алгоритму оптимізації градієнтного спуску. Алгоритм оптимізації градієнтного спуску знаходить градієнт функції втрат.

Сума, за якою оновлюються параметри ваги та зміщення, відома як learning rate. Математичне рівняння для оновлення параметрів ваги та зміщення є наступним:
![title](https://studymachinelearning.com/wp-content/uploads/2019/12/weight_update-2.png)
Градієнт <a style="color:green;">𝝏L/ 𝝏W</a> - часткове виведення функції втрат щодо вагового параметра. Це визначає швидкість зміни похибки щодо змін у ваговому параметрі.

Оновляючи вагові параметри, краще використовувати частку значення градієнта, а не враховувати повну величину.

Наприклад, якщо рівень навчання дорівнює α = 0,1, на кожній ітерації тренування вагові параметри оновлюються лише на 10% від градієнтного спуску.
## Вклад швидкості навчання (Learning Rate) в нейронні мережі
Пошук оптимальної швидкості навчання - важке завдання. **Швидкість навчання (learning rate)** - це параметр настройки, який контролює швидкість, з якою навчається модель.

Занадто висока швидкість навчання дозволяє моделі навчатися швидше, але, можливо, перевищення мінімальної точки, оскільки вага швидко оновлюється.

Невеликий рівень навчання дозволяє моделі навчатися повільно. Це вносить менші зміни до ваги кожного оновлення. Отже, потрібно занадто багато часу для збігання.

Занадто низький рівень навчання може застрягти в небажаному місцевому мінімумі. Тому ми не повинні використовувати рівень навчання занадто великий і занадто малий.

Значення швидкості навчання залежить від архітектури нейронної мережі, а також від нашого набору даних.
## Знаходження оптимальної швидкості навчання
Для досягнення кращих показників роботи вашої нейронної мережі необхідно знайти оптимальне значення швидкості навчання, яке не надто велике і не надто мало. Існує кілька підходів, щоб знайти найкращий рівень навчання.

    Learning Rate Decay
    Learning Rate Schedule
    Adaptive Learning Rate

 

Learning Rate Decay

Learning rate повільно зменшується протягом кожного кроку тренувань, яку називають Learning Rate Decay. Іноді фіксований рівень навчання стикається з труднощами зближення, можливо, через дані або інших факторів. Learning Rate Decay допомагає моделі сходитися за оптимальною величиною.

## Приклади Learning Rate
Розробимо нейронну мережу на наборі ірисів за допомогою бібліотеки навчання Keras Deep. ви можете встановити фіксовану швидкість навчання в класі оптимізатора Кераса. Давайте подивимось, як це працює з датасетом ірисів.

Швидкість навчання визначається в класі оптимізатора Кераса. Нижче наводиться синтаксис для встановлення швидкості навчання 0,1 за допомогою стохастичного оптимізатора спуску градієнта.
```
sgd = optimizers.SGD(lr=0.1)
model.compile(..., optimizer=sgd)
```
Давайте поглянемо на результати при різній швидкості навчання


## import libraries

````python
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets
````

## Load Iris dataset

````python
iris = datasets.load_iris()
X = iris.data
y = iris.target

Y = to_categorical(y)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
````

## Initialize Neural Network

````python
def define_model():
    model = Sequential()
    model.add(Dense(50, input_dim=4, activation='relu'))
    model.add(Dense(12, activation='relu'))
    model.add(Dense(3, activation='softmax'))
    return model
````

## Define Learning rates

````python
l_r = [0.3,0.2,0.1,0.03,0.02,0.01]
````

````python
plt.figure(figsize=(12,9))
for e in range(len(l_r)):
    sgd = optimizers.SGD(lr=l_r[e])
    
    model = define_model()
    model.compile(loss='categorical_crossentropy',optimizer=sgd, metrics=['accuracy'])
    history=model.fit(X_train, y_train,validation_data=(X_test,y_test), epochs=20, batch_size=32)
    
    _, accuracy = model.evaluate(X_test, y_test)
    print('Learning Rate: %f Test Accuracy: %.2f' % (l_r[e],accuracy*100))
    
    plt.subplot(3,3,e+1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Learning Rate: %.2f' % (l_r[e]))
````

## Приклад для Learning Rate Decay

Дефолтне learning rate значення is 0.01. Learning rate decay параметр є 0 по дефолту.
```
sgd = optimizers.SGD(lr=0.1,decay=0.01)
model.compile(..., optimizer=sgd)
```
Давайте поглянемо на приклад

````python
plt.figure(figsize=(12,9))
for e in range(len(l_r)):
    # decay initialize
    sgd = optimizers.SGD(lr=l_r[e], decay=0.01)
    
    model = define_model()
    model.compile(loss='categorical_crossentropy',optimizer=sgd, metrics=['accuracy'])
    history=model.fit(X_train, y_train,validation_data=(X_test,y_test), epochs=20, batch_size=32)
    
    _, accuracy = model.evaluate(X_test, y_test)
    print('Learning Rate: %f Test Accuracy: %.2f' % (l_r[e],accuracy*100))
    
    plt.subplot(3,3,e+1)
    plt.plot(history.history['loss'], label='Training Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.legend(loc='upper right')
    plt.title('Learning Rate: %.2f' % (l_r[e]))
````
